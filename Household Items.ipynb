```python
import requests
import pandas as pd 
import json

# Target URL
target_url = ('http://acnhapi.com/v1/houseware')
```

```python
# Request the data from acnh api
houseware_json = requests.get(target_url).json()

# Test Type for successful request
type(houseware_json)
```

```python
# Explore the data
houseware_json["acoustic_guitar"][0].keys()
```

```python
houseware_json['acoustic_guitar'][0]
```

```python
# Create dictionary with buy price and sell price for each item
price_dict = {}

for item in houseware_json:
    price_dict[str(item)] = {'buy-price': houseware_json[str(item)][0]['buy-price'], 'sell-price': houseware_json[str(item)][0]['sell-price']}
```

```python
price_dict
```

```python
price_df = pd.DataFrame(price_dict).swapaxes("index", "columns")
price_df
```

```python
# Find most expensive household items
price_df.sort_values(['buy-price'], ascending=False).head(25)
```
